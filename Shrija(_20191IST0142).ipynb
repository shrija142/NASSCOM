```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math
```

```python
# importing the dataset
dataset = pd.read_csv('/house_prices7.csv')
pd.set_option('display.max_rows',None)
dataset
```

```python
dataset.info()
```

```python
# independent Variable - Size of the Houses
size_data = dataset['SqFt']
# Target/Dependent Variable - Price of the Houses
price_data = dataset['Price']
size = np.array(size_data).reshape(-1,1)
price = np.array(price_data).reshape(-1,1)
```

```python
# Train the Model
model = LinearRegression()
model.fit(size,price)
```

```python

# Predict Price
price_predicted=model.predict(size)
```

```python
# Plot the result
plt.scatter(size,price, color="green")
plt.plot(size,price_predicted, color="blue")
plt.title("Linear Regression",color="red")
plt.xlabel("House Size")
plt.ylabel("House Price")
plt.show()
```
